```python
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
```

```python
def map_label_to_sentiment(val) -> str:

    if pd.isna(val):
        return np.nan
    
    # Convert to string untuk consistency
    val_str = str(val).lower().strip()

    if val_str.endswith('.0'):
        val_str = val_str[:-2]
    
    if val_str == '1':
        return 'Positive'
    elif val_str == '-1':
        return 'Negative'
    elif val_str == '0':
        return 'Neutral'
    

    if 'positive' in val_str or 'pos' in val_str:
        return 'Positive'
    elif 'negative' in val_str or 'neg' in val_str:
        return 'Negative'
    elif 'neutral' in val_str or 'neut' in val_str:
        return 'Neutral'
    
    return np.nan
```

```python
def prepare_idsmsa() -> pd.DataFrame:
    
    BASE_PATH = r'C:\BINUS\NLP\Research_paper\CS-IDX30 version 3\sentiment analysis'
    path = f"{BASE_PATH}\\IDSMSA.csv"
    
    try:
        df = pd.read_csv(path)
        print(f"   Loaded: {df.shape[0]} records")
        

        df_prep = df[['Sentence', 'Sentiment']].copy()
        df_prep.columns = ['text', 'sentiment']
        df_prep['sentiment'] = df_prep['sentiment'].apply(map_label_to_sentiment)
        df_prep['source'] = 'Twitter (IDSMSA)'
        
        # Filter out NaN labels
        before = len(df_prep)
        df_prep = df_prep.dropna(subset=['sentiment'])
        after = len(df_prep)
        
        print(f"Prepared: {after} records (removed {before-after})")
        print(f"Label dist: {df_prep['sentiment'].value_counts().to_dict()}")
        
        return df_prep
    
    except Exception as e:
        print(f"ERROR: {e}")
        return None
```

```python
def prepare_arto() -> pd.DataFrame:    
    BASE_PATH = r'C:\BINUS\NLP\Research_paper\CS-IDX30 version 3\sentiment analysis'
    path = f"{BASE_PATH}\\dataset_arto_clear_500 comment.xlsx"
    
    try:
        df = pd.read_excel(path, sheet_name='Sheet1')
        print(f"   Loaded: {df.shape[0]} records")
        print(f"   Columns: {list(df.columns)}")
        
        # Explicit columns: ARTO punya 'content_original' dan 'label'
        # Check if columns exist
        if 'content_original' not in df.columns:
            print(f"Column 'content_original' not found!")
            print(f"Available: {list(df.columns)}")
            return None
        
        if 'label' not in df.columns:
            print(f"Column 'label' not found!")
            print(f"Available: {list(df.columns)}")
            return None
        
        df_prep = df[['content_original', 'label']].copy()
        df_prep.columns = ['text', 'sentiment']
        
        # DEBUG: Show sample values sebelum mapping
        print(f"Sample label values: {df_prep['sentiment'].head(3).tolist()}")
        print(f"Label dtype: {df_prep['sentiment'].dtype}")
        
        df_prep['sentiment'] = df_prep['sentiment'].apply(map_label_to_sentiment)
        df_prep['source'] = 'Stockbit (ARTO)'
        
        # Filter out NaN labels
        before = len(df_prep)
        df_prep = df_prep.dropna(subset=['sentiment'])
        after = len(df_prep)
        
        print(f"Prepared: {after} records (removed {before-after})")
        print(f"Label dist: {df_prep['sentiment'].value_counts().to_dict()}")
        
        return df_prep
    
    except Exception as e:
        print(f"ERROR: {e}")
        import traceback
        traceback.print_exc()
        return None
```

```python
def prepare_iccsci() -> pd.DataFrame:
    
    BASE_PATH = r'C:\BINUS\NLP\Research_paper\CS-IDX30 version 3\sentiment analysis'
    path = f"{BASE_PATH}\\dataset_iccsci_clear_1853comment.xlsx"
    
    try:
        df = pd.read_excel(path, sheet_name='Sheet1')
        print(f"   Loaded: {df.shape[0]} records")
        print(f"   Columns: {list(df.columns)}")
        
        if 'content_original' not in df.columns:
            print(f"Column not found!")
            return None
        
        if 'manual' not in df.columns:
            print(f"Column 'manual' not found!")
            print(f"Available columns: {list(df.columns)}")
            return None
        
        label_col = 'manual'
        
        df_prep = df[['content_original', label_col]].copy()
        df_prep.columns = ['text', 'sentiment']
        

        print(f"Sample label values: {df_prep['sentiment'].head(3).tolist()}")
        print(f"Label dtype: {df_prep['sentiment'].dtype}")
        
        df_prep['sentiment'] = df_prep['sentiment'].apply(map_label_to_sentiment)
        df_prep['source'] = 'Stockbit (ICCSCI)'
        
        # Filter out NaN labels
        before = len(df_prep)
        df_prep = df_prep.dropna(subset=['sentiment'])
        after = len(df_prep)
        
        print(f"Prepared: {after} records (removed {before-after})")
        print(f"Label dist: {df_prep['sentiment'].value_counts().to_dict()}")
        
        return df_prep
    
    except Exception as e:
        print(f"ERROR: {e}")
        import traceback
        traceback.print_exc()
        return None
```

```python
def prepare_ptba() -> pd.DataFrame:
    
    BASE_PATH = r'C:\BINUS\NLP\Research_paper\CS-IDX30 version 3\sentiment analysis'
    path = f"{BASE_PATH}\\dataset_ptba_clear_500comment.xlsx"
    
    try:
        df = pd.read_excel(path, sheet_name='Sheet2')
        print(f"   Loaded: {df.shape[0]} records")
        print(f"   Columns: {list(df.columns)}")
        
        # Explicit columns: PTBA punya 'content_original' dan 'Label'
        if 'content_original' not in df.columns:
            print(f"Column 'content_original' not found!")
            return None
        
        if 'Label' not in df.columns:
            print(f"Column 'Label' not found!")
            return None
        
        df_prep = df[['content_original', 'Label']].copy()
        df_prep.columns = ['text', 'sentiment']
        
        # DEBUG: Show sample values sebelum mapping
        print(f"Sample label values: {df_prep['sentiment'].head(3).tolist()}")
        print(f"Label dtype: {df_prep['sentiment'].dtype}")
        
        df_prep['sentiment'] = df_prep['sentiment'].apply(map_label_to_sentiment)
        df_prep['source'] = 'Stockbit (PTBA)'
        
        # Filter out NaN labels
        before = len(df_prep)
        df_prep = df_prep.dropna(subset=['sentiment'])
        after = len(df_prep)
        
        print(f"Prepared: {after} records (removed {before-after})")
        print(f"Label dist: {df_prep['sentiment'].value_counts().to_dict()}")
        
        return df_prep
    
    except Exception as e:
        print(f"   ❌ ERROR: {e}")
        return None
```

```python
def main():
    
    
    idsmsa_prep = prepare_idsmsa()
    arto_prep = prepare_arto()
    iccsci_prep = prepare_iccsci()
    ptba_prep = prepare_ptba()
    
    all_dfs = [
        (idsmsa_prep, 'IDSMSA'),
        (arto_prep, 'ARTO'),
        (iccsci_prep, 'ICCSCI'),
        (ptba_prep, 'PTBA'),
    ]
    
    merged_parts = []
    print("\nFinal counts:")
    for df_prep, name in all_dfs:
        if df_prep is not None and len(df_prep) > 0:
            merged_parts.append(df_prep)
            count = len(df_prep)
            print(f"{name:10}: {count:5} records")
        else:
            print(f"{name:10}: SKIPPED")
    
    if len(merged_parts) == 0:
        print("\nERROR")
        return None
    
    merged_df = pd.concat(merged_parts, ignore_index=True)
    
    print(f"\ntotal merged: {len(merged_df)} records")

    
    print(f"\nSentiment Distribution:")
    for label in ['Positive', 'Negative', 'Neutral']:
        count = (merged_df['sentiment'] == label).sum()
        prop = count / len(merged_df) * 100
        print(f"   {label:12} : {count:5} ({prop:5.1f}%)")
    
    print(f"\nSource Distribution:")
    for source, count in merged_df['source'].value_counts().items():
        prop = count / len(merged_df) * 100
        print(f"   {source:30} : {count:5} ({prop:5.1f}%)")
    
    
    output_path = 'dataset_combined_final.csv'
    merged_df.to_csv(output_path, index=False)
    print(f"\nSaved: {output_path}")
    print(f"Records: {len(merged_df)}")
    

    
    return merged_df
```

```python
merged_df = main()
```
